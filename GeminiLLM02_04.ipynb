```python
from PIL import Image
import google.generativeai as genai
import matplotlib.pyplot as plt

# Configure API key
genai.configure(api_key="AIzaSyDjjVZBCI1PmkKVHNgkJKb09qOF85Wyz9I")

# Use a valid Gemini model
model = genai.GenerativeModel("gemini-1.5-pro")  # Use a listed model

# Process Image Function
def process_image(image_path):
    """Processes an image using the Gemini model.

    Args:
        image_path: The path to the image file.

    Returns:
        The text generated by the Gemini model.
    """
    # Open the image using PIL
    image = Image.open(image_path)

    # Add a text prompt to guide the model
    text_prompt = "Describe this image."

    # Generate content using the image and text prompt
    response = model.generate_content([text_prompt, image])  # Pass image directly

    return response.text, image

# Image Path
image_path = "/content/flower-image-dataset/flowers/bougainvillea_00052.jpg"

# Call the function and get the result
description, image = process_image(image_path)

# Display the image and text
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis("off")
plt.title("Generated Description:")
plt.show()
print("Description:", description)
```

```python
%%writefile app.py
import streamlit as st
import os

st.title("Gemini LLM Multi-Modal AI Hands-On")
st.subheader("Bring Your Own Dataset (Kaggle, Local, APIs)")

# Text Processing
st.write("## Text Processing")
text_input = st.text_area("Enter a prompt:")
if st.button("Generate Text Response"):
    st.write("Placeholder: Generate text response here")

# Image Processing
st.write("## Image Processing")
uploaded_file = st.file_uploader("Upload an Image", type=["jpg", "png", "jpeg"])
if uploaded_file:
    st.image(uploaded_file, caption="Uploaded Image", use_column_width=True)
    if st.button("Analyze Image"):
        st.write("Placeholder: Image analysis output")

# Text + Image Processing
st.write("## Text + Image Processing")
text_prompt_2 = st.text_area("Enter a text prompt for the image:")
if uploaded_file and text_prompt_2 and st.button("Analyze Text + Image"):
    st.write("Placeholder: Text + Image analysis output")

# Kaggle Dataset Integration
st.write("## Load Kaggle Dataset")
dataset_name = st.text_input("Enter Kaggle dataset name:")
if st.button("Download Dataset"):
    os.system(f"kaggle datasets download -d aksha05/flower-image-dataset -p ./data && unzip ./data/flower-image-dataset.zip -d ./data")
    st.write("Downloaded flower-image-dataset successfully!")
```
